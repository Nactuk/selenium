```python
from selenium import webdriver
import time
#selenium ve time import ediliyor
```

```python
def havanasil():
    yer = input("Nerenin hava durumunu öğrenmek istiyorsunuz?... ")
    #hava durumu öğrenilmek istenilen yer string değer olarak alınır
    driver = webdriver.Chrome()
    driver.get("https://www.google.com/")
    #google arama motoru açılır
    aramamotoru = driver.find_element_by_xpath("/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input")
    aramamotoru.send_keys(yer + " hava durumu")
    #arama motoruna hava durumu öğrenilmek istenilen yer + hava durumu diye aratılır. örn:serdivan hava durumu
    aramatusu = driver.find_element_by_xpath("/html/body/div[1]/div[3]/form/div[1]/div[1]/div[3]/center/input[1]")
    aramatusu.click()
    #ara butonuna tıklanır
    time.sleep(6)
    #sayfanın yüklenmesi beklenir
    enyukseksıcaklık = driver.find_element_by_xpath("/html/body/div[7]/div/div[10]/div/div[2]/div[2]/div/div/div[1]/div/div/div/div[3]/div[3]/div/div[1]/div[3]/div[1]/span[1]").text
    endusuksıcaklık = driver.find_element_by_xpath("/html/body/div[7]/div/div[10]/div/div[2]/div[2]/div/div/div[1]/div/div/div/div[3]/div[3]/div/div[1]/div[3]/div[2]/span[1]").text
    #o günün en yüksek ve en düşük sıcaklık değerleri alınır
    yagisoranlari = []
    havadurumlari = []
    gunisimleri = []
    nemoranlari = []
    encoksıcaklık = []
    #verileri içine koyabilmek için diziler tanımlanır
    gunsayisi = 1
    #gün sayısı 1'den başlandı çünkü ilk kutucuğa div[1] ile ulaşılabiliniyordu
    for x in range(8):
    #8 tane kutucuk var
        gunalani = driver.find_element_by_xpath("/html/body/div[7]/div/div[10]/div/div[2]/div[2]/div/div/div[1]/div/div/div/div[3]/div[3]/div/div[" + str(gunsayisi) + "]")
        #gunsayisi adlı değişken ile div[1],div[2],div[3] diye artarak giden butonlara sırasıyla ulaşabiliyoruz
        gunalani.click()
        #her defasında günü değiştiriyoruz
        time.sleep(0.25)
        yagisorani = driver.find_element_by_xpath("/html/body/div[7]/div/div[10]/div/div[2]/div[2]/div/div/div[1]/div/div/div/div[1]/div[2]/div[1]/span").text
        havadurumu = driver.find_element_by_xpath("/html/body/div[7]/div/div[10]/div/div[2]/div[2]/div/div/div[1]/div/div/div/div[2]/span/div[3]/span").text
        gunismi = driver.find_element_by_xpath("/html/body/div[7]/div/div[10]/div/div[2]/div[2]/div/div/div[1]/div/div/div/div[2]/span/div[2]").text
        nemorani = driver.find_element_by_xpath("/html/body/div[7]/div/div[10]/div/div[2]/div[2]/div/div/div[1]/div/div/div/div[1]/div[2]/div[2]/span").text
        sıcaklık = driver.find_element_by_xpath("/html/body/div[7]/div/div[10]/div/div[2]/div[2]/div/div/div[1]/div/div/div/div[1]/div[1]/div/div[1]/span[1]").text
        #her gün için ilgili değerlerin olduğu yerler tekrar tanımlanıyor,değişebilmesi için döngünün içinde
        #istenilen veriler çekildiği için pencere kapatılır
        nemoranlari.append(int(nemorani.split("%")[0]))
        #57% şeklinde bir değer alındığı için split ile bölünüyor ve 57 kısmı alınıyor.ileride karşılaştırma yapabilmek için int değere çevriliyor
        yagisoranlari.append(int(yagisorani.split("%")[0]))
        havadurumlari.append(havadurumu)
        gunisimleri.append(gunismi)
        #karşılaştırmaya ihtiyaç duyulmadığı için alındığı şekilde bırakılıyorlar
        encoksıcaklık.append(int(sıcaklık))
        #direkt 26 gibi sade değer alındığı için sadece int'e çevriliyor
        gunsayisi += 1
        #günler arasında dolaşabilmek için 1 arttırılıyor
    driver.close()
    print(f"\nBugün en çok sıcaklık: {enyukseksıcaklık}°C , en düşük sıcaklık: {endusuksıcaklık}°C , anlık yağış oranı {yagisoranlari[0]}% , hava durumu {havadurumlari[0]}\n")
    #toplanan değerler f string ile kullanıcıya sunuluyor
    enyuksekyagis = 0
    for x in range(len(yagisoranlari)):
    #yagisoranlari listesinin eleman sayısı kadar yani 8 kere döndürülüyor
        if yagisoranlari[x] >= enyuksekyagis:
        #en yüksek değeri bulabilmek için teker teker karrşılaştırılıyorlar.
            enyuksekyagis = yagisoranlari[x]
            #her defasında en yüksek yağış değeri en yüksek yağışı bulana kadar değişir
            enyuksekyagisindex = x
            #en yüksek yağışın olduğu günü seçebilmek için ve o günkü diğer değerlere ulaşabilmek için indexi alınır
    print(f"Yağışın en olası olduğu gün, bu hafta: {gunisimleri[enyuksekyagisindex]} günü , yağış oranı: {yagisoranlari[enyuksekyagisindex]}% , hava durumu: {havadurumlari[enyuksekyagisindex]}\n")
    enyukseknem = 0
    for x in range(len(nemoranlari)):
        if nemoranlari[x] >= enyukseknem:
            enyukseknem = nemoranlari[x]
            enyukseknemindex = x
    print(f"Nemin en olası olduğu gün, bu hafta: {gunisimleri[enyukseknemindex]} günü , nem oranı: {nemoranlari[enyukseknemindex]}% , hava durumu: {havadurumlari[enyukseknemindex]}\n")
    enyukseksıcaklık = 0
    for x in range(len(encoksıcaklık)):
        if encoksıcaklık[x] >= enyukseksıcaklık:
            enyukseksıcaklık = encoksıcaklık[x]
            enyukseksıcaklıkindex = x
    print(f"Sıcaklığın en yüksek olduğu gün, bu hafta: {gunisimleri[enyukseksıcaklıkindex]} günü , sıcaklık: {encoksıcaklık[enyukseksıcaklıkindex]}°C , hava durumu: {havadurumlari[enyukseksıcaklıkindex]}\n")
```

```python
def sarkıac():
    sarkıismi = input("Hangi şarkıyı çalmak istiyorsunuz?...")
    sanatcıismi = input("Şarkıyı kim/kimler seslendiriyor?...")
    #youtubeda aratabilmek için veriler alınır
    driver = webdriver.Chrome()
    driver.get(f"https://www.youtube.com/results?search_query={sanatcıismi} {sarkıismi}")
    #arama butonuna tıklanılmış gibi linke gidilir
    videoyeri = driver.find_element_by_xpath("/html/body/ytd-app/div[1]/ytd-page-manager/ytd-search/div[1]/ytd-two-column-search-results-renderer/div/ytd-section-list-renderer/div[2]/ytd-item-section-renderer/div[3]/ytd-video-renderer[1]/div[1]/div/div[1]/div/h3/a")
    videoyeri.click()
    #videoya tıklanır
```

```python
def farkveyuzdehesapla(farklar,yuzdeler,kurdegerleri,guncel,gun):
    farklar = []
    yuzdeler = []
    for x in range(6):
    #6 adet kur verisi olduğu için döngü 6 kere döndürülür
        farklar.append(round(float(guncel[x]) - float(kurdegerleri[x]),2))
        #güncel kur bilgisinden eski kur bilgisi çıkartılır ve ondalıklı kısmı 2 basamak olacak şekilde yuvarlanır
        yuzdeler.append(str(round(farklar[x] / float(kurdegerleri[x])*100,2)) + "%")
        #değişimin formülü (yeni-eski)/eski
        #yüzdelik değere çevirmek için 100 ile çarpılır ve ondalıklı kısmı 2 basamak olacak şekilde yuvarlanır
    print(f"\n{gun}\nDolar alış farkı: {farklar[0]}, yüzdesi: {yuzdeler[0]}\nDolar satış farkı: {farklar[1]}, yüzdesi: {yuzdeler[1]}\nEuro alış farkı: {farklar[2]}, yüzdesi: {yuzdeler[2]}\nEuro satış farkı: {farklar[3]}, yüzdesi: {yuzdeler[3]}\nİngiliz sterlini alış farkı: {farklar[4]}, yüzdesi: {yuzdeler[4]}\nİngiliz sterlini satış farkı: {farklar[5]}, yüzdesi: {yuzdeler[5]}")        
```

```python
def kurbilgisi():
    driver = webdriver.Chrome()
    driver.get("https://www.tcmb.gov.tr/wps/wcm/connect/tr/tcmb+tr/main+page+site+area/bugun")
    driver.implicitly_wait(15)
    dolaralis = driver.find_element_by_xpath("/html/body/main/div/div[2]/div/div/table[1]/tbody/tr[1]/td[4]").text
    dolarsatis = driver.find_element_by_xpath("/html/body/main/div/div[2]/div/div/table[1]/tbody/tr[1]/td[5]").text
    euroalis = driver.find_element_by_xpath("/html/body/main/div/div[2]/div/div/table[1]/tbody/tr[4]/td[4]").text
    eurosatis = driver.find_element_by_xpath("/html/body/main/div/div[2]/div/div/table[1]/tbody/tr[4]/td[5]").text
    ingsterlinalis = driver.find_element_by_xpath("/html/body/main/div/div[2]/div/div/table[1]/tbody/tr[5]/td[4]").text
    ingsterlinsatis = driver.find_element_by_xpath("/html/body/main/div/div[2]/div/div/table[1]/tbody/tr[5]/td[5]").text
    driver.close()
    guncel = [dolaralis,dolarsatis,euroalis,eurosatis,ingsterlinalis,ingsterlinsatis]
    print(f"\nDolar alış: {dolaralis}\nDolar satış: {dolarsatis}\nEuro alış: {euroalis}\nEuro satış: {eurosatis}\nSterlin alış: {ingsterlinalis}\nSterlin satış: {ingsterlinsatis}")
    bir_altı_yirmiiki = ["16.4127","16.4423","17.5916","17.6233","20.6236","20.7311"]
    yirmialtı_bes_yirmiiki = ["16.3681","16.3975","17.5101","17.5417","20.5722","20.6794"]
    bes_bes_yirmiiki = ["14.8313","14.8580","15.7054","15.7337","18.5428","18.6394"]
    uc_bir_yirmiiki = ["13.4221","13.4463","15.2342","15.2616","18.1055","18.1999"]
    dort_bir_yirmibir = ["7.3602","7.3734","9.0382","9.0545","10.0544","10.1069"]
    iki_bir_yirmi = ["5.9478","5.9585","6.6630","6.6750","7.8464","7.8873"]
    #eski günler için kur değerleri
    print("\n01.06.22 için 1, 26.05.22 için 2, 26.05.22 için 3, 04.01.21 için 4, 02.01.20 için 5")
    cevap = ""
    while cevap != "0":
        farklar = []
        yuzdeler = []
        cevap = input("Hangi gün ile karşılaştırmak istersiniz ?(İstemiyorsanız 0 girin)")
        if cevap == "1":
            gun = "01.06.22"
            farkveyuzdehesapla(farklar,yuzdeler,bir_altı_yirmiiki,guncel,gun)
        elif cevap == "2":
            gun = "26.05.22"
            farkveyuzdehesapla(farklar,yuzdeler,yirmialtı_bes_yirmiiki,guncel,gun)
        elif cevap == "3":
            gun = "05.05.22"
            farkveyuzdehesapla(farklar,yuzdeler,bes_bes_yirmiiki,guncel,gun)
        elif cevap == "4":
            gun = "03.01.21"
            farkveyuzdehesapla(farklar,yuzdeler,uc_bir_yirmiiki,guncel,gun)
        elif cevap == "5":
            gun = "02.01.20"
            farkveyuzdehesapla(farklar,yuzdeler,iki_bir_yirmi,guncel,gun)
        elif cevap == "0":
            continue
        else:
            print("Lütfen istenilen değerlerden birini giriniz!")
```
